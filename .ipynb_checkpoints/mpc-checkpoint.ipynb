# Matrixing  the data 

```python
data.replace(to_replace="Canda",value="Canada",inplace=True)
```

```python
count=0
cmap={}
imap={}
for country in data['Origin']:
    if cmap.get(country)==None:
        cmap[country]=count
        imap[count]=country
        count+=1
        
for country in data['Destination']:
    if cmap.get(country)==None:
        cmap[country]=count
        imap[count]=country
        count+=1
```

```python
import sys
INF=sys.maxsize
rows, cols = (30, 30)
arr = [[INF]*cols]*rows
print("Given Array , len::", len(arr))
for i in arr:
    for j in i:
        if(j==INF):
            print("INF",end=" ")
        else:
            print(j,end=" ")
            
    print()
```

```python
for i in range(len(data)):
    start=cmap.get(data['Origin'][i])
    end=cmap.get(data['Destination'][i])
    if data['Distance(km)'][i]=='NAN': 
        arr[start][end]=INF
    else:
        arr[start][end]=float(data['Distance(km)'][i])

        
print("Indexed Array Created, len::", len(arr))
```

```python
print("                            Indexed Array , len::", len(arr))
print("                       ================================================")
print()
for i in range(len(arr)):
    print("Current Row: " , i, " Cols Len:" , len(arr[i])) 
    for j in range(len(arr[i])):
        if(arr[i][j]==INF):
            print("INF",end=" ")
        else:
            print(arr[i][j],end=" ")
            
    print()
```

# Algorithm

```python
def DFS(s, d, vis, cp, all_paths, cost):
        vis[s]=True
        cp.append(s)
        
        last_added=all_paths[len(all_paths)-1][1]
#         if(cost>last_added):
#             return
        if(s==d):
            print("")
            print("Reached Destination V2.0 with a cost of", cost)
            print(cp)
            for i in range(len(cp)):
                if i !=  len(cp)-1 :
                    print( imap.get(cp[i]), end =  " -> ")
                else:
                    print( imap.get(cp[i]),end =" ")
            print()
            temp=[i for i in cp]
            all_paths.append(temp)
            vis[s]=False
            cp.pop()
            return
    
        for i in range(len(arr)):
            if(vis[i]!=True and arr[s][i]!=INF):
                DFS(i, d, vis , cp, all_paths, cost+arr[s][i])

        vis[s]=False
        cp.pop()
        return
    
```

```python
def getMin(aps):
    minc=INF
    ans=[]
    for c in aps:
        cost=0
        for i in range(1,len(c)-1):
            cost+=arr[c[i]][c[i+1]]    
            if(cost<minc):
                minc=cost
                ans=c

    print("The Min Cost: " , minc)
    print("The Min Cost Path: " , ans)
    for i in ans:
        print(imap.get(i),end=" -> ")
    print()
```

```python
def optimize(source,destination):

    o=cmap.get(source)
    d=cmap.get(destination)
    if(d==None):
        print("No optimal path could be found")
        return []
    
    print("Searching for path b/w u", source,o, " -> " , destination,d)
    vis=[]
    all_paths=[]
    all_paths.append([[],100000000])
    for i in range(len(arr)):
        vis.append(False)
    
    DFS(o,d,vis,[],all_paths,0)
    return all_paths
```
